```python
import pandas as pd
```

```python
data_frame = pd.read_csv('/home/e-saakidis/Documents/bazaraki_car_prices/DataCleaning/car_semi_data_cleaned.csv')
data_frame.head()
```

```python
data_frame.dtypes
```

```python
#conver price, year and km type into int in order to proceed with the analysis
#starting with price need to erase dot and euro symbol first

remove_euro_dot = data_frame['Price'].replace('[\€\.]','',regex=True)
remove_euro_dot.head()
```

```python
remove_price_space = remove_euro_dot.replace('                    ','',regex=True)
remove_price_space.head()
```

```python
#price_int = remove_price_space.astype(int)
```

```python
#need to remove 'from ' & '           New to order' before converting column price to int

price_add = remove_price_space.replace('           New to order','',regex=True)
price_add.head()
```

```python
price_add = remove_price_space.replace('from ','',regex=True)
price_add.head()
```

```python
#price_add.astype(int)
```

```python
price_order_remove = price_add.str.replace('           New to order','',regex=True)
price_order_remove
```

```python
price_order_remove.loc[price_order_remove.str.len() > 15].reset_index(drop=True)
```

```python
price_order_remove.str.strip()
```

```python
price_order_remove.astype(int) #later will join this table with the new km and year int column
```

```python
km_numeric = pd.to_numeric(data_frame['Km'], errors='coerce').astype('Int64')
km_numeric #join later 
```

```python
year_numeric = pd.to_numeric(data_frame['Year'], errors='coerce').astype('Int64')
year_numeric #join
```

```python
#also engine, let's remove the L and make it float
remove_litre = data_frame['Engine'].str.replace('L','',regex=True)
remove_litre
```

```python
engine_replace = remove_litre.replace(',','.',regex=True)
engine_replace
```

```python
engine_replace.loc[engine_replace.str.len() > 3]
```

```python
remove_tric = engine_replace.str.replace('tric','',regex=True)
remove_tric
```

```python
engine_replace.groupby(engine_replace).count()
```

```python
remove_tric.dropna()
```

```python
engine_nan = remove_tric.replace('','NaN',regex=True)
engine_nan
```

```python
engine_nan.dropna()
```

```python
engine_float = engine_nan.astype(float)
engine_float #merge
```

```python
#let's merge the tables
data_merge = pd.merge(data_frame, price_order_remove, left_index=True, right_index=True)
data_merge.head()
```

```python
#data_merge.drop(columns='Price_x', inplace=True).rename(columns={'Price_y':'Price'})
#data_merge
```

```python
data_merge.rename(columns={'Price_y':'Price'}, inplace=True)
data_merge
```

```python
merge_km = pd.merge(data_merge, km_numeric, left_index=True, right_index=True)
merge_km.head()
```

```python
merge_km.rename(columns={'Km_y':'Km'}, inplace=True)
merge_km.head(  )
```

```python
merge_km.drop(columns='Km_x', inplace=True)
merge_km.head()
```

```python
merge_year = pd.merge(merge_km, year_numeric, left_index=True, right_index=True)
merge_year.head()
```

```python
merge_year.rename(columns={'Year_y':'Year'}, inplace=True)
merge_year.drop(columns='Year_x', inplace=True)
merge_year.head()
```

```python
merge_engine = pd.merge(merge_year, engine_float, left_index=True, right_index=True)
merge_engine.rename(columns={'Engine_y':'Engine'}, inplace=True)
merge_engine.drop(columns='Engine_x', inplace=True)
merge_engine.head()
```
