## Split the columns so we have a better "view"

I plan to improve the organization of our records by restructuring the "car" and "km_transmission_gas" column. Specifically, I will separate the car brand, model, engine and year into their own columns. This will facilitate future analysis and provide more detailed information about the cars in our records.

```python
dublicates_remove['Brand'] = dublicates_remove['Car'].str.split().str[0] #Car_Brand
dublicates_remove['Year'] = dublicates_remove['Car'].str[-4:] #Car_Year
dublicates_remove
```

```python
for _, row in dublicates_remove.iterrows():
    words = row['Car'].split()
    dublicates_remove.loc[_, 'Car'] = ' '.join(words[1:-1])
dublicates_remove
```

```python
columns_split = dublicates_remove.copy()
```

```python
data_imported = columns_split.copy()
```

```python
data_imported
```

```python
data_imported['New_Car'] = data_imported['Car'].str[:-4]
data_imported
```

```python
data_imported['Engine'] = data_imported['Car'].str[-4:]
data_imported
```

```python
data_splited = data_imported.copy()
```

```python
data_imported['Price'] = data_imported['Price'].replace(' \n          \n            \n            ','', regex=True)
data_imported
```

```python
def remove_second_euro(x):
    index = x.find('€', x.find('€') + 1)
    x = x[:index] + x[index + 3:]
    return x
```

```python
data_imported['Price'] = data_imported['Price'].apply(remove_second_euro)
data_imported
```

```python
data_dirty = data_splited.copy()
```

```python
data_splited
```

```python
data_splited['Price'] = data_splited['Price'].replace(' \n          \n            \n            ','', regex=True)
data_splited
```

```python
data_splited["Price"] = data_splited["Price"].str.replace("\n", "").str.strip()
data_splited
```

```python
def split_and_join(x):
    splitted = x.split('€')
    return '€'.join(splitted[0:2])

# Apply the function to the 'price' column
data_splited['Price'] = data_splited['Price'].apply(split_and_join)
```

```python
data_splited
```

```python
data_splited['Km'] = data_splited['Km_Transmission_Gas'].str.split().str[0]
data_splited['Transmission'] = data_splited['Km_Transmission_Gas'].str.split().str[2]
data_splited['Gas'] = data_splited['Km_Transmission_Gas'].str.split().str[3]
data_splited
```

```python
data_splited.drop(columns=["Car","Km_Transmission_Gas"], inplace=True) #delete extra columns
data_splited.head()
```

```python
data_splited['Date_Place'] = data_splited['Date_Place'].str.split().str[2]
data_splited.head()
```

```python
#remove the comma from the transmission column
data_splited['Transmission'] = data_splited['Transmission'].replace(",","",regex=True) 
data_splited.head()
```

```python
data_splited.to_csv('car_data_cleaned.csv', index=False)
```
