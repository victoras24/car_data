## Split the columns so we have a better "view"

I plan to improve the organization of our records by restructuring the "car" and "km_transmission_gas" column. Specifically, I will separate the car brand, model, engine and year into their own columns. This will facilitate future analysis and provide more detailed information about the cars in our records.

```python
dublicates_remove['Car_Brand'] = dublicates_remove['Car'].str.split().str[0] #Car_Brand
dublicates_remove['Car_Model'] = dublicates_remove['Car'].str.split(" ", n=2, expand=True)[1] #Car_Model
dublicates_remove['Car_Engine'] = dublicates_remove['Car'].str.split(" ", n=3, expand=True)[2] #Car_Engine
dublicates_remove['Car_Year'] = dublicates_remove['Car'].str[-4:] #Car_Year
dublicates_remove['Km'] = dublicates_remove['Km_Transmission_Gas'].str.split().str[0]
dublicates_remove['Transmission'] = dublicates_remove['Km_Transmission_Gas'].str.split().str[2]
dublicates_remove['Gas'] = dublicates_remove['Km_Transmission_Gas'].str.split().str[3]
dublicates_remove.head()
```

```python
unique_engine = dublicates_remove['Car_Engine'].unique()
unique_engine
```

```python
move_list = ['e-Power','Defender','Range','Discovery','S','Tributo','Mach-E','Astra','Corsa','Star', 'Corsa','Insignia','Stelvio', 'Evolution','Cruiser','Spider','Vectra', 'Spur', 'X','Phantom','Multispace','Giulietta','Antara','Cross','595', 'Country','Vitara','Cherokee','Techno','Mokka','Punto', 'Fe', 'CC','Zafira','GTB','Ghost','Vantage','M','Equilibre', 'Essential', 'Expression','Comfort', 'D3', 'MiTo', 'Aircross', 'DB11', 'GTC', 'Picasso','Levin', 'Wraith', 'Silver', 'Freelander', 'Acenta','Acenta-Comfort', 'Brera', 'Giulia','C-MAX','Adam', 'DBX', 'Meriva', 'II','N-Connecta','Tekna-Premium', 'N-Connecta-Dynamic', 'N-Design', 'Tekna+','E-TECH','Prestige', 'TECHNO', 'Extreme']
for i, row in dublicates_remove.iterrows():
    # Check if the value in column 2 is in the list of search strings
    if row['Car_Engine'] in move_list:
        # If it is, move the value from column 2 to column 1
        dublicates_remove.at[i, 'Car_Model'] = row['Car_Engine']
        dublicates_remove.at[i, 'Car_Engine'] = None
dublicates_remove
```
