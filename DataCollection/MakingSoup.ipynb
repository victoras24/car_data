```python
import pandas as pd
from bs4 import BeautifulSoup
import os
```

```python
titles = []
Description = []
Price = []
Date_Place = []

for file_number in range(2, 72):

    with open(f"/home/sgouros/Desktop/Test/WebScrab3/BazarakiHtml/test{file_number}.html", "r") as reader:
        html_string = reader.read()
        soup = BeautifulSoup(html_string, "html.parser")

    for box in soup.find_all(class_="list-announcement-block"):
        titles.append(box.find_all(class_="announcement-block__title")[0].text.strip())
        Description.append(box.find_all(class_="announcement-block__description")[0].text.strip())
        Price.append(box.find_all(class_="announcement-block-link announcement-block__link")[0].text.strip())
        Date_Place.append(box.find_all(class_="announcement-block__date")[0].text.strip())
    
```

```python
table = pd.DataFrame(data={"Car":titles, "Km_Transmission_Gas":Description, "Price":Price, "Date_Place":Date_Place})
```
